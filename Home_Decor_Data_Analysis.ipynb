```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df=pd.read_csv("home_decor_dataset.csv")
df.head(10)
```

```python
df = df.drop(columns=['Product Name'])
```

```python
df.head()
```

```python
print("\nMissing Values:\n", df.isnull().sum())
```

```python
df.size
```

```python
df.shape
```

```python
df.drop_duplicates(inplace=True)
```

```python
df["Price (USD)"] = pd.to_numeric(df["Price (USD)"], errors="coerce")
df["Rating"] = pd.to_numeric(df["Rating"], errors="coerce")
df["Reviews Count"] = pd.to_numeric(df["Reviews Count"], errors="coerce")
```

```python
df.head()
```

```python
print("\nData Summary:\n", df.describe())
```

```python
top_categories = df["Category"].value_counts().head(5)
print("\nTop 5 Categories:\n", top_categories)
```

```python
top_brands = df["Brand"].value_counts().head(5)
print("\nTop 5 Brands:\n", top_brands)
```

```python
avg_rating_by_category = df.groupby("Category")["Rating"].mean().sort_values(ascending=False)
print("\nAverage Rating by Category:\n", avg_rating_by_category)
```

```python
correlation_matrix = df[["Price (USD)", "Rating", "Reviews Count"]].corr()
print("\nCorrelation Matrix:\n", correlation_matrix)
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df["Price (USD)"], bins=30, kde=True, color="green")
plt.title("Price Distribution of Home Decor Products")
plt.xlabel("Price (USD)")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x=top_categories.index, y=top_categories.values, palette="coolwarm")
plt.title("Top Selling Categories")
plt.xlabel("Category")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(18, 12))
sns.countplot(x="Rating", data=df, palette="Blues")
plt.title("Distribution of Product Ratings")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x=top_brands.index, y=top_brands.values, palette="cividis")
plt.title("Top Brands by Product Count")
plt.xlabel("Brand")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
fig = px.scatter(df, x="Price (USD)", y="Purchase Intent (%)", color="Category", title="Price vs. Purchase Intent")
fig.show()
```
